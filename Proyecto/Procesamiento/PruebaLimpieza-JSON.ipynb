### Prueba de limpieza del CSV y transformación a JSON

```python
import pandas as pd
```

```python
data = pd.read_csv('estaciones_combinadas_NO_LIMPIO.csv', decimal=",")
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data = data.dropna(axis=1, thresh=0.5*len(data)*0.5)
```

```python
data.replace("-", pd.NA, inplace=True)
```

```python
data
```

```python
columnas_importantes =  [
    'COD_ESTACION', 'NOM_ESTACION', 'FECHA', 'HORA', 'SO2', 'CO', 'NO', 'NO2',
    'NOx', 'O3', 'PST', 'Ni', 'Pb'
]
```

```python
data = data[columnas_importantes]
```

```python
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis=1, thresh=0.5*len(data)*0.5)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data["SO2"].isna().sum()
```

```python
columnas_con__nulos = ['SO2', 'NO', 'NO2', 'NOx', 'O3']
```

```python
for columna in columnas_con__nulos:
    data[columna] = pd.to_numeric(data[columna], errors='coerce')
```

```python
for columna in columnas_con__nulos:
    data[columna].fillna(data[columna].mean(), inplace=True)
```

```python
data
```

```python
data.drop("CO", axis=1, inplace=True)
```

```python
import json
```

```python
json_object = data.to_json(orient='records')

with open('estaciones_combinadas_LIMPIO.json', 'w') as file:
    file.write(json_object)
```
