### Prueba de limpieza partiendo el dataframe por Estaci√≥n

```python
import pandas as pd
```

```python
dataCompleto = pd.read_csv('../Ingesta/ESTACIONES_COMBINADAS_NO_LIMPIO/estaciones_combinadas_NO_LIMPIO.csv', decimal=",")
dataCompleto
```

```python
dataElche = dataCompleto[dataCompleto["NOM_ESTACION"] == "ELX - AGROALIMENTARI"]
dataTorrevieja = dataCompleto[dataCompleto["NOM_ESTACION"] == "TORREVIEJA"]
dataOrihuela = dataCompleto[dataCompleto["NOM_ESTACION"] == "ORIHUELA"]
dataAlicante = dataCompleto[dataCompleto["NOM_ESTACION"] == "ALACANT - EL PLA"]
```

```python
def limpiezaAutomaticaPorEstacion(dataFrame):
    dataFrame.replace("-", pd.NA, inplace=True)
    columnasImportantes = [
        'COD_ESTACION', 'NOM_ESTACION', 'FECHA', 'HORA', 'SO2', 'CO', 'NO', 'NO2',
    'NOx', 'O3',
    ]
    
    dataFrame = dataFrame[columnasImportantes]
    
    columnas_con_nulos = ['SO2', 'NO', 'NO2', 'NOx', 'O3']
    
    for columna in columnas_con_nulos:
            dataFrame[columna] = pd.to_numeric(dataFrame[columna], errors='coerce')
            
    for columna in columnas_con_nulos:
        dataFrame[columna].fillna(dataFrame[columna].mean(), inplace=True)
        
    dataFrame = dataFrame.dropna(axis = 1, thresh = 0.5 * len(dataFrame) * 0.5)
        
    return dataFrame
```

```python
limpiezaAutomaticaPorEstacion(dataElche)
```

```python
limpiezaAutomaticaPorEstacion(dataTorrevieja)
```

```python
dataTorrevieja
```

```python
conjuntoDatos = [dataElche, dataTorrevieja, dataOrihuela]
```

```python
for dataFrame in conjuntoDatos:
    limpiezaAutomaticaPorEstacion(dataFrame)
```

```python
import json
```

```python
def dataFrametoJSON(dataFrame, nombre):
    json_object = dataFrame.to_json(orient='records')
    filename = f"{nombre}-LIMPIO.json"
    with open(filename, 'w') as file:
        file.write(json_object)
    print(f"Archivo guardado como {filename}")
```

```python
dataFrametoJSON(dataElche, "Elche")
dataFrametoJSON(dataOrihuela, "Orihuela")
dataFrametoJSON(dataTorrevieja, "Torrevieja")
dataFrametoJSON(dataAlicante, "Alicante")
```

```python
dataElche.to_csv("Elche-Limpio.csv", sep=";", decimal=",")
dataOrihuela.to_csv("Orihuela-Limpio.csv", sep=";", decimal=",")
dataTorrevieja.to_csv("Torrevieja-Limpio.csv", sep=";", decimal=",")
dataAlicante.to_csv("Alicante-Limpio.csv", sep=";", decimal=",")
```
